```python
# ==============================================
# Mathmatical Calculations
import numpy as np
# ==============================================
# Remove Warning
import warnings
warnings.filterwarnings('ignore')
# ==============================================
# GUI APP
from tkinter import *
from tkinter import ttk 
# ==============================================
# Read Image
from PIL import Image
# ==============================================
# Load Saved Model & Scaler
import pickle
import joblib
```

# ===============================================

```python
class Application(Tk):
    def __init__(self):
        Tk.__init__(self)
        self.state('zoomed')
        self.title('Online Payments Fraud Detection Application')
        self.config(bg = '#e9ecef')
        # ============================================================================================
        # ============================================================================================ 
        # ============================================================================================
        # Variable 
        Feature_1 = StringVar()
        Feature_2 = StringVar()
        Feature_3 = StringVar()
        Feature_4 = StringVar()
        Feature_5 = StringVar()
        Feature_6 = StringVar()    
        
        # ======================================================
        # ML Model
        def Detect():
            self.label_entry_8.configure(text = '')
            f1 = Feature_1.get()
            f2 = float(Feature_2.get())
            f3 = float(Feature_3.get())
            f4 = float(Feature_4.get())
            f5 = float(Feature_5.get())
            f6 = float(Feature_6.get())
            # ======================================================
            # First List The Feature
            features = [[f1, f2, f3,f4,f5,f6]]
            # ======================================================
            # ======================================================
            # ======================================================
            # ==> Load LabelEncoder
            with open('label_encoder.pkl', 'rb') as f:
                lab2 = pickle.load(f)
            features[0][0] = lab2.transform(np.array([[features[0][0]]]))
            features[0][0]=features[0][0][0]
            print('features = ',features)
            # ======================================================
            # ======================================================
            # ==> Load RobustScaler For Scaling New Data
            scaler_newData = joblib.load('scaler.joblib')
            x_newData_scaled = scaler_newData.transform(np.array(features))
            print('x_newData_scaled = ',x_newData_scaled)
            # ======================================================
            # ======================================================
            # load the model
            load_model = pickle.load(open("Online Payments Fraud Detection.sav", 'rb'))
            y_newData_ = load_model.predict(x_newData_scaled)
            print('y_newData_ = ',y_newData_[0])
            # Label 
            label_entry_7 = Label(self, text = 'Result Of Detection Using { NN Algorithm }', font=('Arial',30,'bold'),bg = '#e9ecef')
            label_entry_7.place(relx=0.7, rely=0.5, anchor='center') 
            # label 
            ## Result 
            if y_newData_[0] == 0 :
                self.label_entry_8.configure(text = 'No Fraud')
            else:
                self.label_entry_8.configure(text = 'Fraud')
                
            self.label_entry_8.place(relx=0.7, rely=0.6, anchor='center') 
            
                
            
        # Text For Page
        # ============================================================================================
        # ============================================================================================
        Text_title = Label(self, text="Features In Online Payments Fraud ",font=('Arial',35,'bold')
                          ,bg = '#e9ecef')
        Text_title.place(relx=0.5, rely=0.1, anchor='center')
        # ==================================================================================================================
        # ============================================================================================
        #  =============== Feature Type Transaction ===================== 
        # Label 
        label_entry_1 = Label(self, text = 'Type Transaction :', font=('Arial',15,'bold'),bg = '#e9ecef')
        label_entry_1.place(relx=0.08, rely=0.3, anchor='center') 
        # Combobox creation 
        type_feature = ttk.Combobox(self, width = 22,font=('Arial',15,'bold') ,textvariable = Feature_1) 
        # Adding combobox drop down list 
        type_feature['values'] = ('PAYMENT',  'TRANSFER', 'CASH_OUT', 'CASH_IN', 'DEBIT') 
        type_feature.place(relx=0.3, rely=0.3, anchor='center') 
        # ============================================================================================
        # ============================================================================================
        #  =============== Feature Amount  ===================== 
        # Label 
        label_entry_2 = Label(self, text = '      Amount :', font=('Arial',15,'bold'),bg = '#e9ecef')
        label_entry_2.place(relx=0.08, rely=0.4, anchor=E) 
        # entry_2
        entry_2 = Entry(self,justify=CENTER,selectbackground='BLUE',width = 24, font=('Arial',15,'bold'),bg = '#F5F3F4',textvariable = Feature_2)
        entry_2.place(relx=0.3, rely=0.4, anchor='center') 
        # ============================================================================================
        # ============================================================================================
        #  =============== Feature Old Balance Orginal  ===================== 
        # Label 
        label_entry_3 = Label(self, text = ' Old Balance Orginal :', font=('Arial',15,'bold'),bg = '#e9ecef')
        label_entry_3.place(relx=0.085, rely=0.5, anchor='center') 
        # entry_2
        entry_3 = Entry(self,justify=CENTER,selectbackground='BLUE',width = 24, font=('Arial',15,'bold'),bg = '#F5F3F4',textvariable = Feature_3)
        entry_3.place(relx=0.3, rely=0.5, anchor='center') 
        # ============================================================================================
        # ============================================================================================
        #  =============== Feature New Balance Orginal  ===================== 
        # Label 
        label_entry_4 = Label(self, text = 'New Balance Orginal :', font=('Arial',15,'bold'),bg = '#e9ecef')
        label_entry_4.place(relx=0.09, rely=0.6, anchor='center') 
        # Entry 4
        entry_4 = Entry(self,justify=CENTER,selectbackground='BLUE',width = 24, font=('Arial',15,'bold'),bg = '#F5F3F4',textvariable = Feature_4)
        entry_4.place(relx=0.3, rely=0.6, anchor='center') 
        # ============================================================================================
        # ============================================================================================
        #  =============== Feature Old Balance Destination  ===================== 
        # Label 
        label_entry_5 = Label(self, text = 'Old Balance Destination :', font=('Arial',15,'bold'),bg = '#e9ecef')
        label_entry_5.place(relx=0.1, rely=0.7, anchor='center') 
        # Entry 5
        entry_5 = Entry(self,justify=CENTER,selectbackground='BLUE',width = 24, font=('Arial',15,'bold'),bg = '#F5F3F4',textvariable = Feature_5)
        entry_5.place(relx=0.3, rely=0.7 , anchor='center') 
        # ============================================================================================
        # ============================================================================================
        #  =============== Feature New Balance Destination  ===================== 
        # Label 
        label_entry_6 = Label(self, text = 'New Balance Destination :', font=('Arial',15,'bold'),bg = '#e9ecef')
        label_entry_6.place(relx=0.1, rely=0.8, anchor='center') 
        # Entry 6
        entry_6 = Entry(self,justify=CENTER,selectbackground='BLUE',width = 24, font=('Arial',15,'bold'),bg = '#F5F3F4',textvariable = Feature_6)
        entry_6.place(relx=0.3, rely=0.8, anchor='center') 
        # ============================================================================================
        # ============================================================================================
        # ===> Label & Button For Result
        # Button 
        b2 = Button(self, text='Detect', bg='#2E81DF', fg='#fff',
              font=('Arial',20,'bold'),activebackground='#fff',width = 20,
            highlightthickness=0,command=Detect)
        b2.place(relx=0.2, rely=0.94, anchor='center')
        # label
        self.label_entry_8 = Label(self, text = None , font=('Arial',30,'bold'),bg = '#e9ecef')

        
        
        
        
if __name__ == "__main__":
    ui = Application()
    ui.mainloop()
```

# Control Pages

```python
def click_Button():
    window.destroy()
    #Go-To-Application.
    ui = Application()
    ui.mainloop()
```

```python
window = Tk()
# Full Screen Window
window.attributes('-fullscreen', True)
window.title('Online Payments Fraud Detection Application')

window.config(highlightthickness=0)

##
width = 1550  # Width.
height = 1000  # Height.
##

#Create Image Background.
image = Image.open('fraud-detection-payments.jpg')
image = image.resize((1550, 1000))
image = ImageTk.PhotoImage(image)

canvas = Canvas(width=width,height=height,highlightthickness=0)
canvas.create_image(width/2,height/2,image=image)
canvas.pack()
#Button.
b1 = Button(window, text='Continue', bg='#2E81DF', fg='#fff',
              font=('Arial',30,'bold'),activebackground='#fff',width = 25,
            highlightthickness=0,command=click_Button)
b1.place(relx=0.5, rely=0.94, anchor='center')
window.mainloop()
```
